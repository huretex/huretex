#### Deep Learning System - Unreadable Model

```python
dls = SimpleCDLS4MNIST()
dls.train_model()
```

```python
filters_conv_1 = [3]
print(filters_conv_1)
filters_conv_2 = list(range(dls.n_filters_conv_2))
print(filters_conv_2)
```

#### Sequential Information System (SIS)

```python
dls.calculate_activations()
dls.calculate_artifact_clusters()
```

```python
sis = dls.get_sequential_information_system()
print(sis)
```

```python
sis['conv1'] = sis['conv1'].apply(lambda x: x.split('_')[filters_conv_1[0]])
print(sis)
```

#### Rough Set Flow Graph (RSFG) - Readable Twin

```python
rsfg = get_rsfg_df(sis)
print(rsfg)
```

```python
rsfg_layers, node_dicts = get_rsfg_layers(rsfg)
print(rsfg_layers)
print(node_dicts)
```

#### Best Path Flow Matrix (BPFM)

```python
rsfg_matrix = np.zeros((10, dls.n_clusters_conv_1))
print(rsfg_matrix)
```

```python
varbounds = []

for d in node_dicts:

    varbounds.append([0,len(d)-1])

for c1 in range(dls.n_clusters_conv_1):
    for c2 in range(10):

        varbounds[0] = [c1,c1]
        varbounds[3] = [c2,c2]

        best_path_info, best_solution = get_best_confident_path(rsfg_layers, node_dicts, dls.n_clusters_conv_1, varbounds)

        rsfg_matrix[int(best_path_info['target'][2]), int(best_path_info['source'][0])] = best_solution
        
        file_name = 'Filter_'+str('_'.join(str(x) for x in filters_conv_1))+'_source_'+best_path_info['source'][0]+'_target_'+best_path_info['target'][2]

        generate_path_visualisation(file_name, best_path_info, filters_conv_1, filters_conv_2, dls.activations, dls.artifact_clusters)


```

```python
print(rsfg_matrix)
np.savetxt("rsfg_matrix_filter_"+str('_'.join(str(x) for x in filters_conv_1))+".csv", rsfg_matrix, delimiter=";")
```

#### Principal Component Analysis (PCA)

```python
X_pca, evs = perform_pca(rsfg_matrix, info_percentage=0.8)
print("Eigenvectors")
print(evs)
np.savetxt("eigenvectors_filter_"+str('_'.join(str(x) for x in filters_conv_1))+".csv", evs, delimiter=";")
```

#### Intuitionistic Fuzzy Set Equation (IFSE)

```python
A = rsfg_matrix
print('A')
print(A)
b = np.zeros(10)
b[9] = 1
print('b')
print(b)
u = greatest_solution(A, b, Lukasiewicz)
print('u')
print(u)
l = lowest_solution(A, b, u, Lukasiewicz)
print('l')
print(l)
```
