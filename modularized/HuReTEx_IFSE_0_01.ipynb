#### T-norms

```python
def T_M(x, y):
    return min(x, y)


def T_P(x, y):
    return x * y


def T_L(x, y):
    return max(x + y - 1, 0)


def T_F(x, y):
    if x + y > 1:
        return min(x, y)
    else:
        return 0

def T_E(x, y):
    return (x * y)/(2 - x - y + x * y)
```

#### Implications

```python
def implication_min(x, y): #Godela
    if x <= y:
        return 1
    else:
        return y

def dual_implication_max(x, y):
    if x>=y:
        return 0
    else:
        return y

#----------------------------
def implication_TP(x, y): #Goguena
    if x<=y:
        return 1
    else:
        return y/x


def dual_implication_SP(x, y):
    if x>=y:
        return 0
    else:
        return (y-x)/(1-x)

#--------------------

def Lukasiewicz(x, y):
    return min(1 - x + y, 1)


def Fodora(x, y):
    if x <= y:
        return 1
    else:
        return max(1 - x, y)

def Einsteina(x, y):
    if x <= y:
        return 1
    else:
        return (2 * y - x * y)/(x + y - x * y)
```

#### Greatest Solution

```python
def greatest_solution(A, b, implication):
    """
    Computing u - the greatest solution of Ax=b
    u = min_i implication(A(i, j), b(i)).

    Parameters:
        A (numpy.ndarray): Matrix  (m, n).
        b (numpy.ndarray): Vector (m, 1).

    Return:
        numpy.ndarray: Vector u (n, 1).
    """

    A = np.array(A)
    b = np.array(b)
    m = A.shape[0]
    if m != b.shape[0]:
        raise ValueError("The number of rows of matrix A must be equal to the length of vector b.")

    # Minimum for each A(i, j) and b(i)
    n = A.shape[1]
    u = np.ones(n)
    for j in range(n):
        for i in range(m):
            u[j] = min(u[j], implication(A[i,j], b[i])) #np.min(u[j], implication(A[i,j], b[i,0]))

    return u
```
